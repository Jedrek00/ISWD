```python
import os
import sys
import pandas as pd

sys.path.append(os.path.abspath('../'))
from mcda.electre.concordance import *
from mcda.electre.discordance import *
from mcda.electre.credibility import *
from mcda.electre.outranking.ranking import *
from mcda.electre.plot import *
```

```python
indexes = ['Działo przeciwpancerne PaK 43 kal. 88 mm',
           'Działo przeciwpancerne MK I kal. 76.2 mm',
           'Armata przeciwpancerna kal. 57 mm',
           'Działo przeciwpancerne M5 kal. 76.2 mm',
           'Działo przeciwpancerne PaK 35/36 kal. 37 mm',
           'Działo przeciwpancerne PaK 38 kal. 50 mm',
           'Działo przeciwpancerne PaK 40 kal. 75 mm',
           'Armata przeciwpancerna wz. 1937 kal. 47 mm',
           'Działo przeciwpancerne wz. 1937 kal. 45 mm',
           'Działo przeciwpancerne M35 kal. 47 mm',
           'Armata przeciwpancerna Hotchkiss wz. 1934 kal. 25 mm',
           'Działo przeciwpancerne kal. 40 mm',
           'Armata dywizyjna wz. 1942 ZiS-3 76.2 mm',
           'Działo przeciwpancerne M3 kal. 37 mm',
           'Działo przeciwpancerne ZiS wz. 1943 kal. 57 mm']
data = pd.read_csv("dane.csv", header=None)
```

```python
data.index = indexes
data.head()
```

```python
scales  = pd.Series([QuantitativeScale(0, 10000000, PreferenceDirection.MIN),
           QuantitativeScale(0, 100000000), 
           QuantitativeScale(0, 10000000),
           QuantitativeScale(0, 2000000000000),
           QuantitativeScale(0, 360)])
weights = pd.Series([7, 5, 8, 4, 2])
indifference = pd.Series([
    Threshold(0, 300),
    Threshold(0, 2),
    Threshold(0, 50),
    Threshold(0, 1500),
    Threshold(0, 3)
])
preference = pd.Series([
    Threshold(0, 600),
    Threshold(0, 4),
    Threshold(0, 100),
    Threshold(0, 3000),
    Threshold(0, 6)
])

veto = pd.Series([
    Threshold(0, 2000), 
    Threshold(0, 10),
    Threshold(0, 400),
    Threshold(0, 10000),
    Threshold(0, 15)
])
```

```python
concordance_artillery = concordance(
    data, scales, weights, indifference, preference
)
concordance_artillery
```

```python
marginals = discordance_marginals(data, scales, preference, veto)
```

```python
credibility_matrix = credibility_comprehensive(concordance_artillery, non_discordance(marginals, NonDiscordanceType.DC, concordance_artillery))
credibility_matrix
```

```python
downward_order = distillation(credibility_matrix)
downward_order
```

```python
upward_order = distillation(credibility_matrix, upward_order=True)
upward_order
```

```python
final_ranking = final_ranking_matrix(downward_order, upward_order)
final_ranking
```

```python
plot = plot_outranking(final_ranking)
plot
```

```python
ranks_ranking = ranks(final_ranking)
counter = 0
for x in ranks_ranking:
    counter += 1
    print(counter, x)
ranks_ranking
```

```python
median_order_ranking = median_order(ranks_ranking, downward_order, upward_order)
median_order_ranking
```
